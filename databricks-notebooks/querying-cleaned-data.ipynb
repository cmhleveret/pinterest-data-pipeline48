```
# Find the most popular Pinterest category people post to based on their country.

# Your query should return a DataFrame that contains the following columns:
# country
# category
# category_count, a new column containing the desired query output
```

```
joined_df = cleaned_post_data.join(cleaned_geolocation_data, cleaned_post_data.ind == cleaned_geolocation_data.ind, 'inner')
# drop duplicate rows
df_distinct = joined_df.distinct()
display(df_distinct)
```

```
from pyspark.sql.functions import col, count, row_number
from pyspark.sql.window import Window

# group by country and category - count no. of posts in same category & country 
category_counts = df_distinct.groupBy('country', 'category').agg(count('*').alias('category_count'))

# filter out category counts = 1
# category_counts_filtered = category_counts.filter(col('category_count') > 1)

# order table descending
most_popular_category = category_counts.orderBy('category_count', ascending=False)

# define a window specification which partitions by country and orders by category_count in descending order
windowSpec = Window.partitionBy('country').orderBy(col('category_count').desc())

# use the row_number function over the window specification to assign ranks to each row within each country partition
ranked_category_counts = most_popular_category.withColumn("rank", row_number().over(windowSpec))

# filter to keep only the top category (rank = 1) for each country
most_popular_category = ranked_category_counts.filter(col('rank') == 1).drop('rank')

display(most_popular_category)
```

### For each country find the user with the most followers.
- Your query should return a DataFrame that contains the following columns:
- country
- poster_name
- follower_count


```
# join geo and user tables
joined_geo_user_df = cleaned_user_data.join(cleaned_geolocation_data, cleaned_user_data.ind == cleaned_geolocation_data.ind, 'inner')
# drop duplicate ind col
joined_geo_user_df = joined_geo_user_df.drop(cleaned_geolocation_data['ind'])
# join geo_user and post tables
joined_geo_user_post_df = joined_geo_user_df.join(cleaned_post_data, joined_geo_user_df.ind == cleaned_post_data.ind, 'inner')
# drop duplicate ind col
joined_geo_user_post_df = joined_geo_user_post_df.drop(cleaned_geolocation_data['ind'])
# drop duplicate rows
df_distinct = joined_geo_user_post_df.distinct().dropDuplicates()
# select appropriate cols
df_selected = df_distinct.select('country', 'poster_name', 'follower_count')

# define a window specification which partitions by country and orders by category_count in descending order
windowSpec = Window.partitionBy('country').orderBy(col('follower_count').desc())

# use the row_number function over the window specification to assign ranks to each row within each country partition
ranked_follower_counts_by_country = df_selected.withColumn("rank", row_number().over(windowSpec))

# filter to keep only the top category (rank = 1) for each country
most_popular_user_per_country = ranked_follower_counts_by_country.filter(col('rank') == 1).drop('rank')

display(most_popular_user_per_country)
```

### What is the most popular category people post to based on the following age groups:
- 18-24
- 25-35
- 36-50
- +50
### Your query should return a DataFrame that contains the following columns:
- age_group, a new column based on the original age column
- category
- category_count, a new column containing the desired query output

```
df_distinct = joined_geo_user_post_df.distinct().dropDuplicates()
df_selected = df_distinct.select('age', 'category')

# add new column with age groups 
df_grouped = df_selected.withColumn(
    'age_group',
    when((col('age') >= 18) & (col('age') <= 24), '18-24')
    .when((col('age') >= 25) & (col('age') <= 35), '25-35')
    .when((col('age') >= 36) & (col('age') <= 50), '36-50')
    .when((col('age') > 50), '50+')
)

#group by age_group and category
df_category_count = df_grouped.groupBy('age_group', 'category').agg(count('*').alias('category_count'))

# define a window specification which partitions by country and orders by category_count in descending order
windowSpec = Window.partitionBy('age_group').orderBy(col('category_count').desc())

# use the row_number function over the window specification to assign ranks to each row within each country partition
top_category_by_age_group = df_category_count.withColumn("rank", row_number().over(windowSpec))

# filter to keep only the top category (rank = 1) for each country
top_category_by_age_group = top_category_by_age_group.filter(col('rank') == 1).drop('rank')

display(top_category_by_age_group)

```

### Find the median follower count of users that have joined between 2015 and 2020, based on which age group they are part of.

### Your query should return a DataFrame that contains the following columns:
- age_group, a new column based on the original age column
- post_year, a new column that contains only the year from the timestamp column
- median_follower_count, a new column containing the desired query output

```
joined_geo_user_post_df = joined_geo_user_post_df.drop(cleaned_geolocation_data['ind'])
df_distinct = joined_geo_user_post_df.distinct().dropDuplicates()

df_with_year = df_distinct.withColumn("year_joined", year("date_joined"))

df_selected = df_with_year.select('age', 'follower_count', 'year_joined')

df_selected = df_selected.filter((col('year_joined') >= 2015) & (col('year_joined') <= 2020))

# add new column with age groups 
df_grouped = df_selected.withColumn(
    'age_group',
    when((col('age') >= 18) & (col('age') <= 24), '18-24')
    .when((col('age') >= 25) & (col('age') <= 35), '25-35')
    .when((col('age') >= 36) & (col('age') <= 50), '36-50')
    .when((col('age') > 50), '50+')
)

df_meidan = df_grouped.groupBy("year_joined", "age_group").agg(expr("percentile_approx(follower_count, 0.5)").alias("median_follower_count"))

display(df_meidan)
```
